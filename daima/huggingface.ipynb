## 2.加载和使用Tokenizer

### 2.1加载

```python
from transformers import AutoTokenizer

# 加载分词
tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-chinese")
```

```python
type(tokenizer)
```

### 2.2使用

```python
# 1.分词
tokens = tokenizer.tokenize("我爱自然语言处理")
tokens
```

```python
# 2.token -> id
ids = tokenizer.convert_tokens_to_ids(tokens)
ids
```

```python
# 3.id -> token
tokens = tokenizer.convert_ids_to_tokens(ids)
tokens
```

```python
# 4.encode
ids = tokenizer.encode("我爱自然语言处理")
ids
```

```python
# 5.decode
text = tokenizer.decode(ids,skip_special_tokens=True)
text.replace(" ","")
```

### 3.2 查看数据集

### 3.3 预处理数据集

```python
dataset.features
```

```python
# 1.删除列
dataset = dataset.remove_columns(['cat'])
```

```python
dataset.features
```

```python
# 2.过滤行
dataset = dataset.filter(lambda x:x['review'] is not None)
```

```python
# 3.划分数据集
dataset_dict = dataset.train_test_split(test_size=0.2)
dataset_dict
```

```python
dataset_dict['train']
```

```python
# 4.map操作（逐条）

train_dataset = dataset_dict['train']
tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-chinese")

def encode(example):
  # example 结构:{label:1,review:'******'}
  inputs = tokenizer(example['review'],padding='max_length',max_length=128,truncation=True)
  inputs['labels'] = example['label']
  return inputs

train_dataset = train_dataset.map(encode,batched=False,remove_columns=['label','review'])
train_dataset.features
```

```python
# 5.map操作（按批）
test_dataset = dataset_dict['test']
tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-chinese")

def batch_encode(batch):
  inputs = tokenizer(batch['review'],padding='max_length',max_length=128,truncation=True)
  inputs['labels'] = batch['label']
  return inputs

test_dataset = test_dataset.map(batch_encode,batched=True,remove_columns=['label','review'])
test_dataset
```

```python
# 6.map操作（dataset_dict）
test_dataset = dataset_dict['test']
tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-chinese")
def batch_encode(batch):
  inputs = tokenizer(batch['review'],padding='max_length',max_length=128,truncation=True)
  inputs['labels'] = batch['label']
  return inputs
dataset_dict = dataset_dict.map(batch_encode,batched=True,remove_columns=['label','review'])
```
